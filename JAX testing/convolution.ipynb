```python
import jax 
import jax.numpy as jnp
from jax import lax
```

```python
x=jnp.ones((2,10,3))
```

```python
x
```

```python
kernel=jnp.array([[[1.0],[2.0],[1.0]]])
kernel
```

```python
def conv_layer(x,kernel):
    dn=lax.conv_dimension_numbers(x.shape,kernel.shape,('NHC','HIO','NHC'))
    output=lax.conv_general_dilated(x,kernel,(1,),'VALID',dimension_numbers=dn)
    return output
```

```python
print(conv_layer(x,kernel).shape)
```
